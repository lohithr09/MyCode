# parallel operation

```python
data = [24,32,25,9,55.0]
```

```python
rdd1 = sc.parallelize(data)

rdd1.collect()
```

```python
rdd2 = rdd1.map(lambda x: int(x) > 10)
rdd2.collect()
```

```python
rdd2.reduce(lambda x,y: int(x)+ int(y) )
```

# UDF to return the length of the String

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import *
```

```python
def dummy_function(data_str):
    return len(data_str)
```

```python
udfn = udf(dummy_function,StringType())
jsonDF2 = jsonDF1.withColumn("length_project_id",udfn(jsonDF1["_projectid"])).show()
```
